## Relation plot---High level functioning plot
* scatter 
* line

```python
data=pd.DataFrame({'a':[2,4,5,6],'b':[5,7,9,3]})
sb.set()
sb.relplot(x='a',y='b',data=data)
```

```python
help(sb.relplot)
```

```python
sb.relplot(x='a',y='b',hue="a",data=data,size=5)
```
