```python
import seaborn as sb
dir(sb)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.random.rand(20)
y=np.random.rand(20)
plt.plot(x,y,'g*',linewidth=2,markersize=20)
```

```python
help(sb.set)
```

```python
sb.set(color_codes=True)                               # for including all colors
plt.plot(x,y)
```

```python
sb.set(palette="dark")                             # different parameters in colors
plt.plot(x,y)
```

## Relation plot---High level functioning plot
* scatter 
* line

```python
data=pd.DataFrame({'a':[2,4,5,6],'b':[5,7,9,3]})
sb.set()
sb.relplot(x='a',y='b',data=data)
```

```python
help(sb.relplot)
```

```python
sb.relplot(x='a',y='b',hue="a",data=data,size=5)
```

```python
# loading datasets
import seaborn as sb
sb.get_dataset_names()
```

```python
ds=sb.load_dataset("tips")
ds
ds.head()
```

```python
sb.relplot(x="total_bill",y="day",data=ds)
```

```python
sb.relplot(x="total_bill",y="day",data=ds,hue="tip")                            
```

```python
sb.relplot(x="total_bill",y="size",data=ds,hue="tip",kind="line")             # kind uesd for particular graph(line,scatter)
```

```python
h=sb.load_dataset("iris")
h
```

```python
sb.relplot(x="sepal_width",y="petal_width",data=h,hue="species",col="species")
```

```python
sb.relplot(x="sepal_width",y="petal_width",data=h,hue="species",col="species",kind="line")
```
