## create a dataframe with students marks of 5 students and 6 sub.and find
#1.total
#2.avg
#3.min marks
#4.max marks
#5.generate null values to some students
#6.count of null values
#7.all subjects graph
#8.scatter plot

```python
d=pd.DataFrame(np.random.rand(5,6),index=['501','502','503','504','505'],columns=['s1','s2','s3','s4','s5','s6'])
d
```

```python
d.iloc[2,3]=np.nan
d.iloc[1,4]=np.nan
d.iloc[4,2]=np.nan
d
```

```python
d.sum(axis=1)
```

```python
d["total"]=d[0:].sum(axis=1)
d
```

```python
d.mean(axis=1)
```

```python
d["avg"]=d[0:].mean(axis=1)
d
```

```python
plt.plot(d)
```

```python
d.isna()
```

```python
d.isna().sum()
```

```python
d.max(axis=1)
```

```python
d.min(axis=1)
```
