### creating subplot using numpy with sine

```python
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,'bo')
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
plt.show()
```

```python
t=np.arange(0.0,0.5,0.1)
s=np.exp(-t)*np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,label="1st graph with sign wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
plt.show()
t[0]=np.nan
t[1]=np.nan
plt.subplot(2,1,2)
plt.plot(t,s,label="2nd graph with sign wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
plt.show()
```
