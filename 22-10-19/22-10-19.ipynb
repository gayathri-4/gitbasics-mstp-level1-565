```python
import matplotlib.pyplot as p
import numpy as np 
x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50] 
y = [1, 4, 3, 2, 7, 6, 9, 8, 10, 5] 
p.plot(x, y, 'b') 
p.xlabel('x') 
p.ylabel('y') 
p.show() 
```

```python
import matplotlib.pyplot as p
import numpy as np 
p.title("TICKS")
p.xlabel('x') 
p.ylabel('y') 
p.xticks(np.arange(0, 51, 5)) 
p.yticks(np.arange(0, 11, 1)) 
p.plot(x, y, 'b') 
p.show() 
```

# Representing a sign wave
* A=sine(wt)
* A=Amplitude
* W=frequency
* t=time

```python
import numpy as np
import matplotlib.pyplot as p
time=np.arange(1,10,0.1)
Amplitude=np.sin(time)
p.plot(time,Amplitude)
p.title("Sign waves")
p.xlabel("time")
p.ylabel("Amplitude")
p.show()
```

```python
import pandas as pd
df=pd.read_csv("income.csv")
df
```

```python
import matplotlib.pyplot as pl
years=df.columns[2:]
alaska_income=df.iloc[1,2:]
pl.plot(years,alaska_income)
pl.title("income of alaska")
pl.xlabel("years")
pl.ylabel("income")
pl.savefig("alaska income.jpg")            #----> to save figure,plots
pl.show()
```

```python
# plot for all states income from 2009-2013
years=df.columns[-5:]
all_states=df.iloc[0:,2:]
pl.plot(years,all_states)
pl.title("income of all states")
pl.xlabel("years")
pl.ylabel("income")
pl.savefig("all states income from 2009-2013")
pl.show()
```
