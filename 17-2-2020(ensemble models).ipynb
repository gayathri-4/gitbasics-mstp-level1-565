## Ensemble models: 
#### types are:
    * Randomforest  
    * votting calssifier--->Soft voting and hard voting
    * bagging classifier
    * boosting classifier--->adaboostclassifier,gradient boosting classifier

* Voting: building multiple models and simple statistics are used to combine predictions
* Bagging: building multiple models from different subsamples of the training dataset(decision tree)
* Boosting: building multiple models each of which learns to fix the predictions error of a prior model in the chain

## Voting Classifier

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
```

```python
df=load_iris()
df
```

```python
df.feature_names
```

```python
iris=pd.DataFrame(df.data,columns=df.feature_names)
iris
```

```python
x=iris.iloc[:,0:4].values
x
```

```python
y=df.target
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
l_clf=LogisticRegression()
r_clf=RandomForestClassifier()
g_clf=GaussianNB()
```

```python
v_clf=VotingClassifier(estimators=[("lr",l_clf),("rf",r_clf),("gb",g_clf)])
```

```python
v_clf
```

```python
## finding accuracy
## predict(test,predict)
for i in(l_clf,r_clf,g_clf,v_clf):
    i.fit(xtrain,ytrain)
    prediction=i.predict(xtest)
    print(i.__class__,__name__,accuracy_score(ytest,prediction))
```

## Bagging Classifier

```python
# for decision tree 
d=pd.read_csv("diabetes.csv")
d
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
x=d.drop("Outcome",axis=1)
x
```

```python
y=d["Outcome"].values
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.25,random_state=1)
```

```python
b_clf=BaggingClassifier(DecisionTreeClassifier(),bootstrap=True,n_estimators=10,oob_score=True)
b_clf
```

```python
b_clf.fit(xtrain,ytrain)
```

```python
b_clf.oob_score_
```

```python
y_pred=b_clf.predict(xtest)
y_pred
```

```python
print(accuracy_score(ytest,y_pred))
```

#### A/C to oob evalution ,the Bagging classifier is likely to achieve more accuracy on the test

### Adaboost classifier
---> predictind the errors and boosting the values

```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
bc=load_breast_cancer()
bc
```

```python
x=pd.DataFrame(bc.data,columns=bc.feature_names)
x
```

```python
y=pd.Categorical.from_codes(bc.target,bc.target_names)
y
```

```python
encoder=LabelEncoder()
binary_y=pd.Series(encoder.fit_transform(y))
binary_y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,binary_y,test_size=0.25,random_state=1)
```

```python
ad_clf=AdaBoostClassifier(DecisionTreeClassifier(max_depth=2),n_estimators=20)
```

```python
ad_clf.fit(xtrain,ytrain)
```

```python
y_pred=ad_clf.predict(xtest)
y_pred
```

```python
print(accuracy_score(ytest,y_pred))
```

```python
confusion_matrix(ytest,y_pred)
```

## Gradient Boosting Classifier

```python
tr=pd.read_csv("train.csv")
tr
```

```python
ts=pd.read_csv("test.csv")
ts
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
ytrain=tr["Survived"]
#ytrain
```

```python
tr.drop(labels="Survived",axis=1,inplace=True)
```

```python
tr
```

```python
full_df=tr.append(ts)
```

```python
drop_col=["Name","Age","SibSp","Parch","Ticket","Cabin","Embarked"]
```

```python
full_df.drop(labels=drop_col,axis=1,inplace=True)
#full_df
```

```python
full_df.tail()
```

```python
full_df=pd.get_dummies(full_df,columns=["Sex"])
```

```python
full_df.isnull().sum()
```

```python
full_df.fillna(value=0.0,inplace=True)
```

```python
full_df.isnull().sum()
```

```python
xtrain=full_df.values[0:891]
xtrain
xtest=full_df.values[891:]
print(xtest)
```

```python
scaler=MinMaxScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xtrain,ytrain,train_size=0.25,random_state=1)
```

```python
lr=[0.05,0.075,0.1,0.25,0.75,1]
for i in lr:
    gb_clf=GradientBoostingClassifier(n_estimators=20,learning_rate=i,max_features=2)
    gb_clf.fit(xtrain,ytrain)
    print("learning rate",i)
    print("accuracy in training:(0:.3f)".format(gb_clf.score(xtrain,ytrain)))
    print("accuracy in testing:(0:.3f)".format(gb_clf.score(xtrain,ytrain)))
```

```python
gb_clf=GradientBoostingClassifier(n_estimators=20,learning_rate=0.5,max_depth=2,max_features=2)
```

```python
gb_clf.fit(xtrain,ytrain)
```

```python
y_pred=gb_clf.predict(xtest)
y_pred
```

```python
confusion_matrix(ytest,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,y_pred))
```
