```python
from sklearn.datasets import load_boston
boston=load_boston()
boston
```

## steps
* Data calling

* train,test

* predict

```python
boston.data
```

```python
# creating dataset
import pandas as pd
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
boston_df["Land rate"]=boston.target
boston_df.head()
```

```python
x=boston_df.drop("Land rate",axis=1)
x.head()
```

```python
x.values
```

```python
y=boston_df["Land rate"].values
y
```

## Model calling(Ridge regression)
* to overcome of over fitting
* RR=sum of error+sum of sqres of magnitude of coeffecient(alpha)


```python
from sklearn.linear_model import Ridge
rr=Ridge(alpha=1,solver='svd')
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)     #train and test
```

```python
rr.fit(x_train,y_train)              #datafitting
```

```python
y_predict=rr.predict(x_train)                             #predict
y_predict
```

```python
rr.score(x_train,y_train)                                      # score
```

```python
rr.score(x_test,y_test)
```

```python
rr.coef_
```

```python
rr.intercept_
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
train_score = lr.score(x_train,y_train)
print("Linear Regression train score:",train_score)
```

```python
test_score = lr.score(x_test,y_test)
print("Linear Regression test score:",test_score)
```

```python
ridge_train_score = lr.score(x_train,y_train)
print("Ridge Regression train score:",ridge_train_score)
```

```python
ridge_test_score = lr.score(x_test,y_test)
print("Ridge Regression test score:",ridge_test_score)
```

```python
plt.plot(rr.coef_,marker='*',markersize = 5,color='red',label = "Ridge coefficient")
plt.plot(lr.coef_,marker='o',markersize = 10,color='green',label = 'Linear Regression')
plt.xlabel('Coefficient Index')
plt.ylabel('Coefficient index')
plt.title("ridge vs linear")
plt.legend()
plt.show()
```

```python
rr2=Ridge(alpha=100,solver="cholesky")
rr2
```

```python
rr2.fit(x_train,y_train)
```

```python
rr2.coef_
```

```python
plt.plot(rr2.coef_,marker='*',markersize = 5,color='red',label = "Ridge coefficient")
plt.plot(lr.coef_,marker='o',markersize = 10,color='green',label = 'Linear Regression')
plt.xlabel('Coefficient Index')
plt.ylabel('Coefficient index')
plt.title("ridge vs linear")
plt.legend()
plt.show()
```
